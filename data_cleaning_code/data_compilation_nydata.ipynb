```python
# Import required libraries. No other libraries are required for this task.

import pandas as pd
import numpy as np
```

## Data Cleaning for Newyork

```python
#### analyzing 1 file of NY Data

path = "C:/Users/ameen/Desktop/Covid_learning_gap_ML_project_17_oct/datasets/newyork"

df2017_path = path + "/3-8_ELA_AND_MATH_RESEARCHER_FILE_2017.csv" 
df2018_path = path + "/3-8_ELA_AND_MATH_RESEARCHER_FILE_2018.csv"
df2019_path = path + "/3-8_ELA_AND_MATH_RESEARCHER_FILE_2019.csv"
df2021_path = path + "/3-8_ELA_AND_MATH_RESEARCHER_FILE_2021.csv"

df2017 = pd.read_csv(df2017_path)
df2018 = pd.read_csv(df2018_path)
df2019 = pd.read_csv(df2019_path)
df2021 = pd.read_csv(df2021_path)

print(df2017.shape)
print(df2018.shape)
print(df2019.shape)
print(df2021.shape)

## df2021 has three less columns
## some rows dont have data for mean scale score in 2021 file
```

```python
### cutting relevant columns and checking unique values and NAs. 
relevant_col = ["SY_END_DATE","NAME","ITEM_SUBJECT_AREA","ITEM_DESC","SUBGROUP_CODE","SUBGROUP_NAME","MEAN_SCALE_SCORE"]

df2017 = df2017[relevant_col]
df2018 = df2018[relevant_col]
df2019 = df2019[relevant_col]
df2021 = df2021[relevant_col]

df_ny = pd.concat([df2017,df2018,df2019,df2021])

df_ny.head()
```

```python
df = df2017

for col in df:
    print("couunt of NA values in", col)
    print(df[col].isnull().sum())
    
print(df["MEAN_SCALE_SCORE"].value_counts())
```

```python
### remove NAs:
df_ny_fin = df_ny.dropna()
print(df_ny_fin.shape)

for col in df_ny_fin:
    print(df_ny_fin[col].unique())

df_ny_fin.info()
```

```python
df_ny_fin.isnull().values.any()
```

```python
### removing entries with no mean scale score
df_ny_fin_1 = df_ny_fin.query("MEAN_SCALE_SCORE != '-'")
print(df_ny_fin.shape)
print(df_ny_fin_1.shape)                              

df_ny_fin_1["MEAN_SCALE_SCORE"] = df_ny_fin_1["MEAN_SCALE_SCORE"].astype(int)
df_ny_fin_1.info()
```
