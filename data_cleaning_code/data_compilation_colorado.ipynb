```python
# Import required libraries. No other libraries are required for this task.
import pandas as pd
import numpy as np
```

## Data Cleaning Colorado

```python
#### analyzing 1 file of Colorado Data

path = "C:/Users/ameen/Desktop/Covid_learning_gap_ML_project_17_oct/datasets/colorado"

df2017_path = path + "/2017_colorado.xlsx"
df2018_path = path + "/2018_colorado.xlsx"
df2019_path = path + "/2019_colorado.xlsx"
df2021_path = path + "/2021_colorado.xlsx"

df2017 = pd.read_excel(df2017_path)
df2018 = pd.read_excel(df2018_path)
df2019 = pd.read_excel(df2019_path)
df2021 = pd.read_excel(df2021_path)

df2017["year"] = '2017'
df2018["year"] = '2018'
df2019["year"] = '2019'
df2021["year"] = '2021'

###2017 has different naming conventions
df2017 = df2017[['year','Level','District Code','District \nName','School \nName','Content','Test', 'Mean \nScale \nScore']]
```

```python
print(df2017.shape)
print(df2018.shape)
print(df2019.shape)
print(df2021.shape)
```

```python
df2017.columns = np.array(['year','Level','District Code','District Name','School Name','Content','Test/Grade', 'Mean Scale Score']) 

print(df2021.columns)
```

```python
df2017 = df2017[df2017['Test/Grade'].str.contains('Grade')]
df2017['Test/Grade'] = df2017['Test/Grade'].apply(lambda x: int(str(x).split()[2]))
df2017['Content'] = df2017['Content'].apply(lambda x:'english' if (x == 'ELA') else ('maths'))
df2017.head()
```

```python
df2017.groupby(['Content','Test/Grade'])['year'].count()
```

```python
df2018 = df2018[(df2018['Test/Grade'].str.contains('English')) | (df2018['Test/Grade'].str.contains('Mathematics'))] 
df2018['Test/Grade'] = df2018['Test/Grade'].apply(lambda x: (str(x).split()[-1]))
df2018['Content'] = df2018['Content'].apply(lambda x:'maths' if (x == 'Mathematics') else ('english'))
df2018['Test/Grade'] = df2018['Test/Grade'].astype('int') 
df2018.head()
```

```python
df2018.groupby(['Content','Test/Grade'])['year'].count()
```

```python
df2019 = df2019[df2019['Test/Grade'] != 'All Grades']
df2019['Test/Grade'] = df2019['Test/Grade'].astype('int')
df2019['Content'] = df2019['Content'].apply(lambda x:'maths' if (x == 'Mathematics') else ('english'))
#df2019.groupby(['Content','Test/Grade'])['year'].count()
df2019.head()
```

```python
df2019.groupby(['Content','Test/Grade'])['year'].count()
```

```python
df2021 = df2021[(df2021['Content'].str.contains('English')) | (df2021['Content'].str.contains('Math'))]
df2021['Content'] = df2021['Content'].apply(lambda x:'maths' if (x == 'Mathematics') else ('english'))
df2021.groupby(['Content','Test/Grade'])['year'].count()
```

```python
relevant_col = ['year','Level','District Code','District Name', 'School Name','Content', 'Test/Grade','Mean Scale Score']

df2017 = df2017[relevant_col]
df2018 = df2018[relevant_col]
df2019 = df2019[relevant_col]
df2021 = df2021[relevant_col]

df_col = pd.concat([df2017,df2018,df2019,df2021])

df_col.columns = np.array(['year','Level','District_Code','District_Name', 'School_Name','Content', 'Grade','Mean_Scale_Score'])
df_col = df_col[df_col['Level'] == 'SCHOOL']

df_col.head()
```

```python
df_col.groupby(['Content','Grade'])['year'].count()
```

```python
df = df_col

for col in df:
    print("couunt of NA values in", col)
    print(df[col].isnull().sum())
    
    
print(df["Mean_Scale_Score"].value_counts())

##need to remove star
```

```python
### remove NAs:
df_col_fin = df_col.dropna()
print(df_col_fin.shape)

for col in df_col_fin:
    print(df_col_fin[col].unique())
    
df_col_fin.info()
```

```python
df_col_fin.isnull().values.any()
```

```python
### removing entries with no mean scale score
df_col_fin_1 = df_col_fin.query("Mean_Scale_Score != '*' & Mean_Scale_Score != '- -' ")

print(df_col_fin.shape)
print(df_col_fin_1.shape)                              

df_col_fin_1["Mean_Scale_Score"] = df_col_fin_1["Mean_Scale_Score"].astype(int)
df_col_fin_1.info()
```

```python
df_col_fin_1.groupby(['Level','School_Name'])['year'].count()
```

```python
### final table for Colorado data:
#### it contains for each year, for each school, for each grade, for each subject and ethnic group the mean scale scores

df_col_fin_1.to_csv('C:/Users/ameen/Desktop/Covid_learning_gap_ML_project_17_oct/datasets_processed/colorado.csv')
df_col_fin_1.head()
```

## District to School Mapping

```python
df_dist_school_map = df_col_fin_1[['School_Name','District_Name']]
df_dist_school_map['School_Name'] = df_dist_school_map['School_Name'].astype('str').str.lower()
df_dist_school_map['District_Name'] = df_dist_school_map['District_Name'].astype('str').str.lower()



print('shape of old mapping')
print(df_dist_school_map.shape)
print('\n')

df_dist_school_map = df_dist_school_map.drop_duplicates(subset = ['School_Name','District_Name'],keep = 'last',inplace=False, ignore_index=False) 
df_dist_school_map.columns = ['school','district']
df_dist_school_map['state'] = 'colorado'

print('shape of new mapping')
print(df_dist_school_map.shape)
print('\n')

df_dist_school_map.to_csv('C:/Users/ameen/Desktop/Covid_learning_gap_ML_project_17_oct/datasets_processed/colorado_school_dist_mapping.csv')
```
